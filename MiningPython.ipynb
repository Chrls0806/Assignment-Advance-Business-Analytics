**Look at CSV files:** TID,item pairs

```python
# Read shopping dataset from CSV file
# Create dictionary "Sitems" with key = item and value = set of transactions
# Also set variable Snumtrans = number of transactions
Sitems = {}
trans = []  # list of transactions used to set Snumtrans
with open('Shop.csv') as f:
    rows = csv.DictReader(f)
    for r in rows:
        if r['item'] not in Sitems:
            Sitems[r['item']] = {r['TID']}
        else:
            Sitems[r['item']].add(r['TID'])
        if r['TID'] not in trans:
            trans.append(r['TID'])
Snumtrans = len(trans)
print('Number of transactions:', Snumtrans)
print('Number of distinct items:', len(Sitems))
print('Item dictionary:')
Sitems
```

```python
# Read movies dataset from CSV file
# Create dictionary "Mitems" with key = item and value = set of transactions
# Also set variable Mnumtrans = number of transactions
Mitems = {}
trans = []  # list of transactions used to set Mnumtrans
with open('Movies.csv') as f:
    rows = csv.DictReader(f)
    for r in rows:
        if r['item'] not in Mitems:
            Mitems[r['item']] = {r['TID']}
        else:
            Mitems[r['item']].add(r['TID'])
        if r['TID'] not in trans:
            trans.append(r['TID'])
Mnumtrans = len(trans)
print('Number of transactions (users):', Mnumtrans)
print('Number of distinct items (movies):', len(Mitems))
print('Item dictionary:')
Mitems.items()
```

#### Mine for association rules in the Movies dataset with three items on the left-hand side. Find support and confidence thresholds (need not be the same) so the number of association rules is more than 10 but less than 20.

```python
# Frequent item-sets of three
support = 0.03    # adjust
confidence = 0.6  # adjust

freq_itemsets_4 = []

from itertools import combinations # Import combinations here

for i1 in Mitems:
    for i2 in Mitems:
        for i3 in Mitems:
            for i4 in Mitems:
                if i1 < i2 < i3 < i4:   # avoid duplicates
                    common = len(Mitems[i1] & Mitems[i2] & Mitems[i3] & Mitems[i4])
                    supp = common / Mnumtrans
                    if supp >= support:
                        freq_itemsets_4.append((frozenset([i1, i2, i3, i4]), supp))

# Step 2: Generate association rules (3 → 1)
rules = []
for itemset, supp in freq_itemsets_4:
    items = list(itemset)
    for left in combinations(items, 3):   # now works
        right = set(items) - set(left)
        left = frozenset(left)
        right = frozenset(right)

        # support of antecedent
        left_support = len(set.intersection(*[Mitems[i] for i in left])) / Mnumtrans

        # confidence
        conf = supp / left_support if left_support > 0 else 0

        if conf >= confidence:
            rules.append((left, right, supp, conf))

# Step 3: Filter for 10 < num_rules < 20
if 10 < len(rules) < 20:
    print(f" Found {len(rules)} rules (support={support}, confidence={confidence})")
    for left, right, supp, conf in rules:
        print(f"{set(left)} -> {set(right)}, support={supp:.3f}, confidence={conf:.3f}")
else:
    print(f" Number of rules = {len(rules)} (outside 10–20 range). Try different thresholds.")
```

#### Mine for association rules in the Movies dataset with three items on the left-hand side. Find support and lift thresholds so the number of association rules is more than 10 but less than 20. Only consider lift thresholds > 1.


```python
# Association rules with three items on the left-hand side
support = 0.03    # adjust this value
lift_threshold = 1.2   # >1 as required, adjust to get 10–20 rules

freq_itemsets_4 = []

for i1 in Mitems:
    for i2 in Mitems:
        for i3 in Mitems:
            for i4 in Mitems:
                if i1 < i2 < i3 < i4:   # avoid duplicates
                    common = len(Mitems[i1] & Mitems[i2] & Mitems[i3] & Mitems[i4])
                    supp = common / Mnumtrans
                    if supp >= support:
                        freq_itemsets_4.append((frozenset([i1, i2, i3, i4]), supp))

# Step 2: Generate association rules (3 → 1)
rules = []
for itemset, supp in freq_itemsets_4:
    items = list(itemset)
    for left in combinations(items, 3):
        right = set(items) - set(left)
        left = frozenset(left)
        right = frozenset(right)

        # support of antecedent (LHS)
        left_support = len(set.intersection(*[Mitems[i] for i in left])) / Mnumtrans

        # support of consequent (RHS)
        right_support = len(set.intersection(*[Mitems[i] for i in right])) / Mnumtrans

        # confidence
        conf = supp / left_support if left_support > 0 else 0

        # lift
        lift = conf / right_support if right_support > 0 else 0

        if supp >= support and lift > lift_threshold:
            rules.append((left, right, supp, conf, lift))

#  Step 3: Filter for 10 < num_rules < 20
if 10 < len(rules) < 20:
    print(f" Found {len(rules)} rules (support={support}, lift>{lift_threshold})")
    for left, right, supp, conf, lift in rules:
        print(f"{set(left)} -> {set(right)}, "
              f"support={supp:.3f}, confidence={conf:.3f}, lift={lift:.3f}")
else:
    print(f" Number of rules = {len(rules)} (outside 10–20 range). "
          f"Try different thresholds (support={support}, lift>{lift_threshold}).")
```
